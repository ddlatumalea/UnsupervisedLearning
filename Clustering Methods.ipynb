**K-means**:
- class: KMeans
- distance: euclidean
- Use case: find a few clusters of ~ same size
- Remark:
    - fast
    - metric used is Euclidean distance, and variance is a measure of cluster scatter. If clusters have different variances, K-means can fail to cluster them. It expects to be spherical-shaped.
    - Diagnostic tests to determine $k$.
    - converges to a local minimum, therefore run it multiple times `n_iter`.

**Mean shift**:
- class: MeanShift
- Distance: euclidean
- Use case: determine the (large) amount of clusters
- Remark: slow, bad on weird shapes

**HAC**: bottom-up. Each data forms a separate cluster. Distances between all clusters are computed. Closest clusters are merged. Repeated until convergence.
- Linkage criterion: defining the distance between a point and a cluster or between clusters:
    - Complete linkage: distance set to maximum distance between elements of each cluster.
    - Single linnkage: distance set to minimum distance between elements of each cluster.
    - Ward's method (Ward's minimum variance method): minimize the total within-cluster variance.
- Determining clusters:
    - Distance threshold.
    - Specifying n_clusters
- class: AgglomerativeClustering
- distance: L1/L2/cosine/Jaccard/Hamming/precomputed, any pairwise method.
- Use case: get a full tree (informative)
- remark: slow, good on weird shapes or uneven cluster sizes. Visualized in dendrogram. Performance depends on distance metric.

**DBSCAN (density-based spatial clustering of applications with noise)**:
- class: DBSCAN
- Distance: any
- use case: tons of data and weird shapes
- remark: highly configurable with good performance. Uses density. Cannot cluster well when there are large differences in densities. OPTICS algorithm may work.
